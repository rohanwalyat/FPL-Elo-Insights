```python
import pandas as pd

players = pd.read_csv("../data/2024-2025/players/players.csv")
matches = pd.read_csv("../data/2024-2025/matches/matches.csv")
playermatch = pd.read_csv("../data/2024-2025/playermatchstats/playermatchstats.csv")

players.head()
matches[['match_id','gameweek','kickoff_time','home_team_elo','away_team_elo']].head()
playermatch.columns.tolist()
```

```python
# aggregate goals per player
goals = playermatch.groupby('player_id')['goals'].sum().reset_index()
goals = goals.merge(players[['player_id','first_name','second_name','web_name']], on='player_id', how='left')
goals.sort_values('goals', ascending=False).head(15)
```
